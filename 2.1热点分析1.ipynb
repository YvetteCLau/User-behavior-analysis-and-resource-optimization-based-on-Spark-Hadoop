```python
import os
import pandas as pd
import jieba
from collections import Counter
from pyecharts.charts import WordCloud
from pyecharts import options as opts

# **1.  CSV **

csv_path = os.path.join( ".csv")

if not os.path.exists(csv_path):
    raise FileNotFoundError(f" : {csv_path}")

# **2.  CSV **
df = pd.read_csv(csv_path)

# **3.  "" **
df.columns = [col.strip() for col in df.columns]  # 
if "" not in df.columns:
    raise ValueError(f" ''  CSV : {df.columns}")

# **4. **
titles = df[""].dropna().astype(str).tolist()

# **5.  `jieba` **
text = " ".join(titles)
words = jieba.cut(text)

# **6. **
stopwords = set(["", "", "", "", "", "", "", "", "", ""])  # 
word_freq = Counter(word for word in words if word not in stopwords)

# **7. **
word_data = word_freq.most_common(200)  #  200 

# **8.  `pyecharts` **
wordcloud = (
    WordCloud()
    .add("", word_data, word_size_range=[20, 100], shape="circle")  # 
    .set_global_opts(title_opts=opts.TitleOpts(title=" "))
)

# **9.  Jupyter Notebook **
wordcloud.render_notebook()
```
